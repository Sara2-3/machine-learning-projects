### Exploring Yes Responses by Job, Education, Marital, and Contact

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# -------------------- JOB DATA --------------------
job_data = {
    "job": [
        "student","retired","unemployed","admin.","management","unknown",
        "technician","self-employed","housemaid","entrepreneur","services","blue-collar"
    ],
    "Yes Count": [
        275,434,144,1352,328,37,730,149,106,124,323,638
    ],
    "Share of Yes (%)": [
        5.93,9.35,3.10,29.14,7.07,0.80,15.73,3.21,2.28,2.67,6.96,13.75
    ]
}
job_report = pd.DataFrame(job_data)

# -------------------- EDUCATION DATA --------------------
education_data = {
    "education": ["primary", "secondary", "tertiary", "unknown"],
    "Yes Count": [475, 2197, 1845, 119],
    "Share of Yes (%)": [10.25, 47.38, 39.81, 2.56]
}
education_report = pd.DataFrame(education_data).set_index("education")

# -------------------- MARITAL DATA --------------------
marital_data = {
    "marital": ["married", "single", "divorced"],
    "Yes Count": [2750, 1445, 438],
    "Share of Yes (%)": [59.36, 31.18, 9.46]
}
marital_report = pd.DataFrame(marital_data).set_index("marital")

# -------------------- CONTACT DATA --------------------
contact_data = {
    "contact": ["cellular", "telephone", "unknown"],
    "Yes Count": [2896, 737, 0],
    "Share of Yes (%)": [79.7, 20.3, 0.0]
}
contact_report = pd.DataFrame(contact_data).set_index("contact")

# -------------------- FUNCTION TO MAP PERCENTAGE TO COLOR --------------------
def map_to_blue_shades(percentages):
    # Normalizojmë përqindjet në [0,1]
    norm = (percentages - min(percentages)) / (max(percentages) - min(percentages))
    # Sa më e lartë përqindja → më e errët
    return plt.cm.Blues(0.4 + 0.5*norm)

# -------------------- PLOTTING --------------------
fig, axs = plt.subplots(2, 2, figsize=(14, 12))

# Pie 1: Job
colors_job = map_to_blue_shades(job_report["Share of Yes (%)"])
axs[0, 0].pie(
    job_report["Yes Count"],
    labels=[f"{job} ({percent}%)" for job, percent in zip(job_report["job"], job_report["Share of Yes (%)"])],
    startangle=140,
    colors=colors_job
)
axs[0, 0].set_title("Share of Yes (%) by Job")

# Pie 2: Education
colors_edu = map_to_blue_shades(education_report["Share of Yes (%)"])
axs[0, 1].pie(
    education_report["Yes Count"],
    labels=[f"{edu} ({percent}%)" for edu, percent in zip(education_report.index, education_report["Share of Yes (%)"])],
    startangle=140,
    colors=colors_edu
)
axs[0, 1].set_title("Share of Yes (%) by Education")

# Pie 3: Marital
colors_marital = map_to_blue_shades(marital_report["Share of Yes (%)"])
axs[1, 0].pie(
    marital_report["Yes Count"],
    labels=[f"{status} ({percent}%)" for status, percent in zip(marital_report.index, marital_report["Share of Yes (%)"])],
    startangle=140,
    colors=colors_marital
)
axs[1, 0].set_title("Share of Yes (%) by Marital Status")

# Pie 4: Contact
colors_contact = map_to_blue_shades(contact_report["Share of Yes (%)"])
axs[1, 1].pie(
    contact_report["Yes Count"],
    labels=[f"{ct} ({percent}%)" for ct, percent in zip(contact_report.index, contact_report["Share of Yes (%)"])],
    startangle=140,
    colors=colors_contact
)
axs[1, 1].set_title("Share of Yes (%) by Contact Type")

plt.tight_layout()
plt.show()
```

## When do clients subscribe most — by month and weekday?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define order for months and weekdays
month_order = ['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec']
day_order = ['mon','tue','wed','thu','fri']

# Filter only subscribed clients (y == "yes")
df_yes = df[df['y'] == 'yes']

# Group data by month and weekday
counts_yes = df_yes.groupby(['month','day_of_week']).size().reset_index(name='count')

# Pivot to wide format: rows = months, columns = weekdays
pivot_yes = counts_yes.pivot(index='month', columns='day_of_week', values='count').fillna(0)

# Reorder rows and columns
pivot_yes = pivot_yes.reindex(index=month_order, columns=day_order)

# Plot stacked area chart (Streamgraph style) with Blue Shades
pivot_yes.plot.area(figsize=(12,6), cmap='Blues', alpha=0.85)
plt.title("Streamgraph of YES Subscriptions by Month and Weekday", fontsize=16, weight='bold')
plt.xlabel("Month", fontsize=14)
plt.ylabel("Number of YES Subscriptions", fontsize=14)
plt.legend(title="Day of Week")
plt.show()
```

### Summary of Streamgraph Insights

#### Month Dimension
- **Observation:** May has the highest subscriptions, while September has the lowest.  
- **Interpretation:** Campaigns are more effective in spring (May), while September shows reduced responsiveness after summer holidays.  

---

#### Weekday Dimension
- **Observation:** Friday has the highest subscriptions, while Monday has the lowest.  
- **Interpretation:** Clients are more receptive at the end of the week (Friday), while Mondays are busier and less effective for marketing calls.

##  Generations of Trust: Who Subscribes?

```python
import pandas as pd

# Define age bins and labels
bins = [18, 25, 35, 45, 55, 65, 100]
labels = ['18–25', '26–35', '36–45', '46–55', '56–65', '66+']

# Create age group column
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)

# Group by age group and calculate metrics
summary = df.groupby('age_group').agg(
    Total_Clients=('age', 'count'),
    Yes_Subscriptions=('y', lambda x: (x == 'yes').sum())
).reset_index()

# Calculate subscription rate
summary['Subscription Rate (%)'] = (summary['Yes_Subscriptions'] / summary['Total_Clients'] * 100).round(1)

# Display table
summary
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Të dhënat nga tabela jote
data = {
    'age_group': ['18–25','26–35','36–45','46–55','56–65','66+'],
    'Total_Clients': [1063, 13686, 13500, 8704, 3567, 663],
    'Yes_Subscriptions': [254, 1666, 1168, 753, 484, 313]
}

df_summary = pd.DataFrame(data)

# Shndërro në format të gjatë për vizualizim
df_melted = pd.melt(df_summary, id_vars='age_group',
                    value_vars=['Total_Clients','Yes_Subscriptions'],
                    var_name='Type', value_name='Count')

# Vizato grafikun
plt.figure(figsize=(6,6))
sns.barplot(x='age_group', y='Count', hue='Type', data=df_melted, palette='Blues_d')

# Stilizim
plt.title('Total Clients vs Yes Subscriptions by Age Group', fontsize=20, weight='bold')
plt.xlabel('Age Group', fontsize=10)
plt.ylabel('Count', fontsize=10)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=14)
plt.legend(title='Category', fontsize=12, title_fontsize=14)

# Shto numrat mbi shtyllat
for p in plt.gca().patches:
    height = p.get_height()
    if height > 0:  # vetëm nëse ka vlerë
        plt.gca().annotate(f'{int(height)}',
                           (p.get_x() + p.get_width() / 2., height),
                           ha='center', va='bottom', fontsize=12, weight='bold')

plt.tight_layout()
plt.show()
```

##  Impact of Economic Indicators on Client Subscription

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# List of macroeconomic variables from the Bank Marketing dataset
macro_vars = ['emp.var.rate', 'cons.price.idx', 'cons.conf.idx', 'euribor3m', 'nr.employed']

# Set style for seaborn plots
sns.set(style='whitegrid')

# Create a grid of 3 rows × 2 columns (we have 5 variables, so one subplot will be empty)
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(24,20))  # large figure size
axes = axes.flatten()

for i, var in enumerate(macro_vars):
    # Bin each variable into 5 intervals
    df[f'{var}_bin'] = pd.cut(df[var], bins=5)
    
    # Calculate subscription rate (proportion of y = 'yes') for each bin
    rates = df.groupby(f'{var}_bin')['y'].apply(lambda x: (x == 'yes').mean()).reset_index()
    rates.columns = [f'{var}_bin', 'subscription_rate']
    
    # Plot bar chart
    ax = axes[i]
    sns.barplot(x=f'{var}_bin', y='subscription_rate', data=rates, ax=ax, palette='Blues_d')
    
    # Titles and labels with larger fonts
    ax.set_title(f'Subscription Rate vs {var}', fontsize=24, weight='bold')
    ax.set_ylabel('Proportion of y = yes', fontsize=20)
    ax.set_xlabel(var, fontsize=20)
    ax.tick_params(axis='x', rotation=45, labelsize=18)
    ax.tick_params(axis='y', labelsize=18)
    
    # Add annotations above each bar to show exact subscription rate
    for p in ax.patches:
        height = p.get_height()
        ax.annotate(f'{height:.2f}', 
                    (p.get_x() + p.get_width() / 2., height),
                    ha='center', va='bottom', fontsize=18, color='black', weight='bold')

# Remove the last empty subplot (since we only have 5 variables, not 6)
fig.delaxes(axes[-1])

# Adjust spacing and show the plots
plt.tight_layout()
plt.show()
```

### Call Duration vs Campaign Contacts by Contact History (pdays)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Categorize pdays into 3 clear groups
def categorize_pdays(value):
    if value == 999:
        return 'Never Contacted'
    elif value == 0:
        return 'Recently Contacted'
    else:
        return 'Previously Contacted'

df['pdays_group'] = df['pdays'].apply(categorize_pdays)

# Eye-friendly color palette
eye_friendly_palette = {
    'Never Contacted': '#1f77b4',      # calm blue
    'Recently Contacted': '#2ca02c',   # soft green
    'Previously Contacted': '#ff7f0e'  # warm orange
}

# Scatter plot
plt.figure(figsize=(10,6))
sns.scatterplot(
    x='campaign',
    y='duration',
    hue='pdays_group',
    data=df,
    palette=eye_friendly_palette,
    alpha=0.7,
    edgecolor='gray'
)

# Styling
plt.title('Duration vs Campaign (Eye-Friendly Colors by Contact History)', fontsize=18, weight='bold')
plt.xlabel('Campaign (Number of Contacts)', fontsize=14)
plt.ylabel('Duration (Seconds)', fontsize=14)

# Legend improvements
plt.legend(
    title='Contact Status',
    fontsize=12,
    title_fontsize=14,
    loc='upper right',
    frameon=True,
    facecolor='white',
    edgecolor='black'
)

plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

##  Contact History Analysis

### 1. Never Contacted (pdays = 999)
- These clients appear mostly in areas with low campaign counts and varied call durations.  
- Suggests that **first-time contacts can be effective**, especially when the call is longer.  

---

### 2. Recently Contacted (pdays = 0)
- Points cluster around higher campaign counts but shorter call durations.  
- May indicate that **clients contacted very recently are less responsive**, or that agents spend less time on these calls.  

---

### 3. Previously Contacted (pdays between 1 and 998)
- This group is more scattered and doesn’t dominate any particular region.  
- Suggests that **past contact doesn’t guarantee success**, especially if many days have passed.  

---

##  Strategic Summary
- Longer calls and **first-time contact (pdays = 999)** show better potential for success.  
- **Frequent follow-ups (pdays = 0)** may not be as effective, especially when campaign count is high.  
- **Old contacts (pdays > 0)** are less predictable and may need a different strategy.

##  Financial Stability Score vs Subscription

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create the new feature
df['stability_score'] = 100 \
    - 20*(df['default'].map({'yes':1,'no':0,'unknown':0})) \
    - 15*(df['housing'].map({'yes':1,'no':0,'unknown':0})) \
    - 15*(df['loan'].map({'yes':1,'no':0,'unknown':0}))

# Plot violin plot of subscription vs stability score
plt.figure(figsize=(12,6))
sns.violinplot(x='y', y='stability_score', data=df, palette=['#1f77b4', '#ff7f0e'])
plt.title('Client Financial Stability vs Subscription Outcome', fontsize=18, weight='bold')
plt.xlabel('Subscription Outcome (y)', fontsize=14)
plt.ylabel('Financial Stability Score', fontsize=14)
plt.show()

```

##  Key Insights

- Clients with **higher financial stability** are more responsive to marketing campaigns.  
- Those with **defaults or loans** tend to reject the offer.  
- This **engineered feature** provides a strategic advantage — helping to target the right clients more effectively.

## Duration Effect
- Call duration is the single strongest predictor.  
- **Longer calls** (e.g., >300 seconds) show much higher subscription rates.  
- **Short calls** (<200 seconds) rarely lead to “yes.”

## Campaign Effect
- Fewer contacts per client are more effective.  
- **1–3 contacts → higher success.**  
- **Many contacts (>5)** → diminishing returns, often lower subscription probability.  

## pdays Effect
- **Never contacted before (pdays = 999)** → higher responsiveness.  
- **Recently contacted (pdays = 0)** → lower success, possibly due to fatigue.  
- **Previously contacted (>0)** → mixed results, not as strong as first-time contact.  

## Demographics (Age, Job, Marital, Education)
- **Older clients (56+)** show higher subscription rates despite smaller numbers.  
- **Blue-collar and low-education groups** are less likely to subscribe.  
- **Married clients** are more stable but not necessarily more responsive.  
- **Students and retirees** show higher-than-average interest.  

## Contact Type
- **Cellular phone** is far more effective than traditional telephone.  
- Most “yes” outcomes come from cellular contacts.

## Financial Stability Score (Engineered Feature)
- Clients with **high stability scores** (no defaults, fewer loans) → much higher subscription probability.  
- Clients with **low stability scores** (defaults + multiple loans) → rarely subscribe, even with repeated campaigns.  

## Interactions
- **Long duration + first-time contact (pdays = 999)** → best scenario.  
- **Short duration + repeated campaigns** → worst scenario.  
- Combining demographics with call features shows compounding effects  
  (e.g., older clients + long calls = very high success).
- **High score + long call + first‑time contact** → best scenario.  
- **Low score + short call + repeated campaigns** → worst scenario

When you bring all the evidence together, the most important features for predicting subscription are:
1. **Duration** (long calls matter most)  
2. **Campaign** (fewer contacts are better)  
3. **pdays** (first-time contact is strongest)  
4. **Contact type** (cellular > telephone)  
5. **Age group** (older clients more likely to subscribe)

#  Conclusion: Term Deposit Subscription Analysis

Based on the evidence and dashboard results:

- **Call duration** is the strongest driver of success. Long calls combined with first-time contact (pdays = 999) yield the highest subscription probability.  
- **Campaign intensity** matters: fewer, well‑timed contacts are more effective than repeated attempts.  
- **Client profile factors** (age, job, education, marital status) influence responsiveness, with older clients, students, and retirees showing higher interest.  
- **Contact type** is critical: cellular communication far outperforms traditional telephone.  
- **Financial stability** amplifies outcomes: clients with no defaults and fewer loans are significantly more likely to subscribe.  
- **Best scenario:** High stability + long call + first‑time contact → maximum success.  
- **Worst scenario:** Low stability + short call + repeated campaigns → minimal success.  

###  Final Insight
The bank should prioritize **longer, quality calls to financially stable clients who are being contacted for the first time via cellular phone**. This strategy maximizes efficiency and subscription rates, while avoiding wasted effort on repeated short calls to less stable clients.
